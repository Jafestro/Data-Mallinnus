#### Finally, calculate the correlation matrix and visualize it for each data frame. Clearly describe the results and your interpretation for it.

```python
import seaborn as sns

# Select only numeric columns
df_affected_numeric = df_affected.select_dtypes(include=[np.number])
df_control_numeric = df_control.select_dtypes(include=[np.number])

# Calculate the correlation matrix
corr_matrix_affected = df_affected_numeric.corr()
corr_matrix_control = df_control_numeric.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_affected, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Affected Group')
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_control, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Control Group')
plt.show()
```

##### Affected:
 Hemoglobin and Packed Cell Volume (0.95): This strong positive correlation suggests that as hemoglobin levels increase, packed cell volume also tends to increase. This is expected since packed cell volume reflects the concentration of red blood cells, which are heavily influenced by hemoglobin.

##### Control:
Hemoglobin and Packed Cell Volume (0.57): The positive correlation between hemoglobin and packed cell volume is weaker compared to the affected group (0.95). This suggests less dependency between these two parameters in the control group, likely reflecting a healthier physiological balance.
